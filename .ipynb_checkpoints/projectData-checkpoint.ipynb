```python
import pandas as pd

df = pd.read_csv('IBM_data_60_years.csv')
df.head()
```

```python
df = df.drop(columns=['Dividends', 'Stock Splits'])
```

```python
# המרה של עמודת 'Date' לפורמט datetime כולל שימוש ב-UTC
df['Date'] = pd.to_datetime(df['Date'], errors='coerce', utc=True)

# הסרת אזור הזמן
df['Date'] = df['Date'].dt.tz_localize(None)

# נורמליזציה להסרת חלק השעה
df['Date'] = df['Date'].dt.normalize()


```

```python
df.head()
```

```python
df.info()
```

```python
monthly_sequences = df.resample('M', on='Date').mean()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(monthly_sequences.index, monthly_sequences['Close'], label='Monthly Closing Price', color='b')
plt.title('Monthly Closing Prices of IBM Stock')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.grid(True)
plt.legend()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

# יצירת אובייקט ה-Scaler לנרמול הנתונים
scaler = StandardScaler()

# נורמליזציה של הנתונים (ללא עמודת התאריך)
normalized_data = scaler.fit_transform(monthly_sequences)

# הצגת התוצאה המנורמלת (חמשת השורות הראשונות)
normalized_df = pd.DataFrame(normalized_data, columns=monthly_sequences.columns, index=monthly_sequences.index)
normalized_df.head()
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# בחירת שני מאפיינים מהנתונים - למשל 'Close' (מחיר סגירה) ו-'Volume' (נפח מסחר)
X = monthly_sequences[['Close', 'Volume']].values

# הגדרת מספר הקלאסטרים (ננסה תחילה עם 3)
kmeans = KMeans(n_clusters=3, random_state=42)
y_kmeans = kmeans.fit_predict(X)

# ציור הקלאסטרים
plt.figure(figsize=(10, 6))

# קלאסטר 1
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s=50, c='red', label='Cluster 1')

# קלאסטר 2
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s=50, c='blue', label='Cluster 2')

# קלאסטר 3
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s=50, c='green', label='Cluster 3')

# הצגת המרכזים של הקלאסטרים (centroids)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='black', label='Centroids')

# הוספת כותרת וצירים
plt.title('Clusters of IBM Stock Data')
plt.xlabel('Closing Price')
plt.ylabel('Volume')
plt.legend(bbox_to_anchor=[1, 0.8])
plt.grid(True)
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# סינון הנתונים משנת 1983 ואילך
monthly_sequences_filtered = monthly_sequences[monthly_sequences.index >= '1983-01-01']

# שלב 1: נורמליזציה של הנתונים המסוננים
scaler = StandardScaler()
normalized_data_filtered = scaler.fit_transform(monthly_sequences_filtered.drop(columns=['Cluster']))

# שלב 2: הפחתת מימדים ל-2 בעזרת PCA
pca = PCA(n_components=2)
pca_data_filtered = pca.fit_transform(normalized_data_filtered)

# שלב 3: ביצוע K-means עם 5 קלאסטרים על הנתונים המסוננים
kmeans_filtered = KMeans(n_clusters=5, random_state=42)
kmeans_filtered.fit(pca_data_filtered)

# שלב 4: הוספת תוויות הקלאסטרים לנתונים המסוננים
monthly_sequences_filtered['Cluster'] = kmeans_filtered.labels_

# שלב 5: חישוב Silhouette Score
silhouette_avg_filtered = silhouette_score(pca_data_filtered, kmeans_filtered.labels_)
print(f'Silhouette Score: {silhouette_avg_filtered}')

# שלב 6: הצגת החודשים שנמצאים בכל קלאסטר לאחר סינון
months_in_clusters_filtered = monthly_sequences_filtered[['Cluster']].reset_index()

# הצגת הממוצעים של כל קלאסטר עבור הנתונים המסוננים
cluster_summary_filtered = monthly_sequences_filtered.groupby('Cluster').mean()
print(cluster_summary_filtered)

# שלב 7: יצירת גרף פיזור עבור הנתונים המסוננים
plt.figure(figsize=(10, 6))

# ציור כל קלאסטר בצבע שונה
plt.scatter(pca_data_filtered[kmeans_filtered.labels_ == 0, 0], pca_data_filtered[kmeans_filtered.labels_ == 0, 1], s=50, c='red', label='Cluster 1')
plt.scatter(pca_data_filtered[kmeans_filtered.labels_ == 1, 0], pca_data_filtered[kmeans_filtered.labels_ == 1, 1], s=50, c='blue', label='Cluster 2')
plt.scatter(pca_data_filtered[kmeans_filtered.labels_ == 2, 0], pca_data_filtered[kmeans_filtered.labels_ == 2, 1], s=50, c='green', label='Cluster 3')
plt.scatter(pca_data_filtered[kmeans_filtered.labels_ == 3, 0], pca_data_filtered[kmeans_filtered.labels_ == 3, 1], s=50, c='purple', label='Cluster 4')
plt.scatter(pca_data_filtered[kmeans_filtered.labels_ == 4, 0], pca_data_filtered[kmeans_filtered.labels_ == 4, 1], s=50, c='magenta', label='Cluster 5')

# ציור המרכזים של כל קלאסטר
plt.scatter(kmeans_filtered.cluster_centers_[:, 0], kmeans_filtered.cluster_centers_[:, 1], s=200, c='black', label='Centroids')

# כותרת וצירים
plt.title('Clusters of IBM Stock Data (1983 and later)')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend(bbox_to_anchor=[1, 0.8])
plt.grid(True)
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score

# שלב 1: הוספת עמודת תנודתיות (באחוזים)
monthly_sequences_filtered['Volatility (%)'] = ((monthly_sequences_filtered['High'] - monthly_sequences_filtered['Low']) / monthly_sequences_filtered['Low']) * 100

# שלב 2: נורמליזציה של הנתונים (כולל התנודתיות)
scaler = StandardScaler()
normalized_data_with_volatility = scaler.fit_transform(monthly_sequences_filtered)

# שלב 3: ביצוע PCA להפחתת מימדים ל-2 לצורך המחשה גרפית
pca = PCA(n_components=2)
pca_data_with_volatility = pca.fit_transform(normalized_data_with_volatility)

# שלב 4: ביצוע K-means עם 5 קלאסטרים
kmeans_with_volatility = KMeans(n_clusters=5, random_state=42)
kmeans_with_volatility.fit(pca_data_with_volatility)


# שלב 6: חישוב Silhouette Score
silhouette_avg_with_volatility = silhouette_score(pca_data_with_volatility, kmeans_with_volatility.labels_)
print(f'Silhouette Score (with Volatility): {silhouette_avg_with_volatility}')

# שלב 7: יצירת גרף פיזור עבור הנתונים הכוללים תנודתיות
plt.figure(figsize=(10, 6))

# ציור כל קלאסטר בצבע שונה
plt.scatter(pca_data_with_volatility[kmeans_with_volatility.labels_ == 0, 0], pca_data_with_volatility[kmeans_with_volatility.labels_ == 0, 1], s=50, c='red', label='Cluster 1')
plt.scatter(pca_data_with_volatility[kmeans_with_volatility.labels_ == 1, 0], pca_data_with_volatility[kmeans_with_volatility.labels_ == 1, 1], s=50, c='blue', label='Cluster 2')
plt.scatter(pca_data_with_volatility[kmeans_with_volatility.labels_ == 2, 0], pca_data_with_volatility[kmeans_with_volatility.labels_ == 2, 1], s=50, c='green', label='Cluster 3')
plt.scatter(pca_data_with_volatility[kmeans_with_volatility.labels_ == 3, 0], pca_data_with_volatility[kmeans_with_volatility.labels_ == 3, 1], s=50, c='purple', label='Cluster 4')
plt.scatter(pca_data_with_volatility[kmeans_with_volatility.labels_ == 4, 0], pca_data_with_volatility[kmeans_with_volatility.labels_ == 4, 1], s=50, c='magenta', label='Cluster 5')

# ציור המרכזים של כל קלאסטר
plt.scatter(kmeans_with_volatility.cluster_centers_[:, 0], kmeans_with_volatility.cluster_centers_[:, 1], s=200, c='black', label='Centroids')

# הוספת כותרת וצירים
plt.title('Clusters of IBM Stock Data with Volatility (1983 and later)')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend(bbox_to_anchor=[1, 0.8])
plt.grid(True)
plt.show()
```
